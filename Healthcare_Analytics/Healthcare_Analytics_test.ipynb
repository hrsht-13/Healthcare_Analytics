```
std_encoding=df.groupby(['Category1']).agg({'outcome':['std']}).reset_index()
df = df.merge(std_encoding,on='Category1',how="left")
df.rename(columns={('outcome', 'std'):"Category1_std"},inplace=True)
```

```
std_encoding=df.groupby(['Category2']).agg({'outcome':['std']}).reset_index()
df = df.merge(std_encoding,on='Category2',how="left")
df.rename(columns={('outcome', 'std'):"Category2_std"},inplace=True)
```
